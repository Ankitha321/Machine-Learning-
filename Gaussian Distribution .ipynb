```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm 
```

```python
domain = np.linspace(0,10,10)
plt.plot(domain , norm.pdf(domain,5,1))
plt.title('Gaussian Distribution for 10 Samples mean = 5 and var = 1')
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
